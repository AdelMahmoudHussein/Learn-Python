```python
#08# how importing it?

#Importing module from a package
#We can import modules from packages using the dot (.) operator.

#or example, if want to import the start module in the above example, it is done as follows.

import Game.Level.start

#Now if this module contains a function named select_difficulty(),
#we must use the full name to reference it.

Game.Level.start.select_difficulty(2)

#If this construct seems lengthy, we can import the module without the package prefix as follows.

from Game.Level import start
#We can now call the function simply as follows.

start.select_difficulty(2)
#Yet another way of importing just the required function (or class or variable) 
#form a module within a package would be as follows.

from Game.Level.start import select_difficulty
#Now we can directly call this function.

select_difficulty(2)
#Although easier, this method is not recommended. 
#Using the full namespace avoids confusion and prevents two same identifier names from colliding.
```

```python
#2#process & Thread
#python includes sophisticaled tools for managing concurrentoperations using processes athreads
#simple programs can be made to run faster by applying techniques for running parts of the job concurrently using these modules

#process vs thread
what is process ?
- an executing instance of program is caleed process
- a process is always stored in the main memory  or RAM
- a process is termed as an active entity . it disappears if the machine is rebooted
- multi process may be associated with a same program
- ex : executing multiple instance of calculator


what is a thread ?
a thread is a subset of the process
it is termed as light wight process ,since it is similar to a real process but
excute within context of a process and shares the same resources
allocated to the process by kernal

aprocess has only one thread of control - one set of machine instructions
excuting at a time

aprocess may also be made up of multiple thrads of execution that
execute instration concurrently

all the threads running within a process sharethe same address space
,file description ,stack and other process related attributes


#threading module
the newer threading module included with python 2.4 provide much more powerful
,high level support for threads

threading module exposes all the methods of the thread module
threading module has the tread class that implements threading

## thread class

##methods in thread class

run(): entry point of a thread
start():starting thread by calling the run method
join([time]) :waiting for thread terminate
isAlive():checks whether a thread is still executing
getName():return the name of thread
setName():sets the name of thread
```

```python
#example 3
from threading import Thread
import time

def my_loop_10_times(i):
    print("sleeping 5 sec from thread %d"%i)
    time.sleep(5)
    print("finished sleeping 5 sec from thread %d"%i)

for i in range(10):
    Thread(target=my_loop_10_times,args=(i,)).start()
```

```python
#thread name
#each thread instance has a name with a default value that can be 
#changed as the thread is created

#naming threads is useful in server processes made up of multiople service
#threads handling different operations

import time 
import threading

def myfunction():
    print (threading.currentThread().getName(),'stating\n')
    time.sleep(2)
    print (threading.currentThread().getName(),'exiting\n')
    
def myservice():
    print (threading.currentThread().getName(),'stating\n')
    time.sleep(2)
    print (threading.currentThread().getName(),'existing\n')
    
f2=threading.Thread(name='my service' , target=myservice)
f1=threading.Thread(name='my function' , target=myfunction)
f3=threading.Thread(target=myfunction)
f4 = threading.Thread(target=myservice)

f1.start()
f2.start()
f3.start()              #default name 
f4.start()              #default name 
```

```python
#deamon thread
#programs sometimes spawn a thread as adaemon that runs without
#blocking the main program from existing

#using deamon thread is useful for services where there may not be
#an easy way to interrupt the thread or where letting the thread die in the
#middle of its work does not lose or corrupt data

#for example a thread that generates 'heart beats' for service monitoring
#tool

#to mark a thread as daemon call its setDaemon() method with true
#the default for threads to not be daemons

#Example
import time 
import threading

def myfunction():
    print (threading.currentThread().getName(),'stating\n')
    time.sleep(2)
    print (threading.currentThread().getName(),'exiting\n')
    
def myservice():
    print (threading.currentThread().getName(),'stating\n')
    time.sleep(2)
    print (threading.currentThread().getName(),'existing\n')
    
f2=threading.Thread(name='my service' , target=myservice)
f2.setDaemon(True)
f1=threading.Thread(name='my function' , target=myfunction)
f3=threading.Thread(target=myfunction)
#f4 = threading.Thread(target=myservice)

f1.start()
f2.start()
f3.start()              #default name 
#f4.start()              #default name 
```
