```python
#08# how importing it?

#Importing module from a package
#We can import modules from packages using the dot (.) operator.

#or example, if want to import the start module in the above example, it is done as follows.

import Game.Level.start

#Now if this module contains a function named select_difficulty(),
#we must use the full name to reference it.

Game.Level.start.select_difficulty(2)

#If this construct seems lengthy, we can import the module without the package prefix as follows.

from Game.Level import start
#We can now call the function simply as follows.

start.select_difficulty(2)
#Yet another way of importing just the required function (or class or variable) 
#form a module within a package would be as follows.

from Game.Level.start import select_difficulty
#Now we can directly call this function.

select_difficulty(2)
#Although easier, this method is not recommended. 
#Using the full namespace avoids confusion and prevents two same identifier names from colliding.
```

```python
#1# OS Module
#this module contains all methods to deal with OS .. files , dirs ,paths
# deals ith text files

import os

#before you can modify file (read Write ) you have to open it first using built-in method open()
#why need to open file 
#Access Mode
# R         |   open a file for reading only
# R+        |   open a file for reading and Writing
# W         |   open a file for writing only
# W+        |   open a file for reading and Writing   
# A         |   open a file for Appending
# A+        |   open a file for reading and Appending

file = open('test.txt','wb')
#note# with write mode will delete file contains
print('file name :', file.name)
print('file closed or not:', file.closed)
print('file mode :', file.mode)
#run in spider editor

#-----------------------------------------
#read method read()
#-----------------------------------------
file = open('test.txt','r+')
words = file.read(20) # 20 num of Char
print (words)
#if you want read all lines in file
file = open('test.txt','r+')
lines = file.readlines()
print(lines)
#if you want print only first line
file = open('test.txt','r+')
line = file.readline()
print(line)

#--------------------------------------------
#tell me postion use tell() method
#--------------------------------------------
file = open('test.txt','r+')
postition = file.tell()
print(postition)
words = file.read(20)
print(postition) # NO
print(file.tell()) #yes

#----------------------------------------------
#Writing in file use Write method
#----------------------------------------------
file = open ('test2.txt' , 'w')
file.write('iam here')
#it will write this in file but not save it so must close() file to save
file.close() #important

#----------------------------------------------
#to Append file add text to file
#----------------------------------------------
file = open ('test2.txt' , 'a')
file.write('lazzem tt3ab')
file.close() #important

#-------------------------------------
# to rename file use rename() method take two argument (current file name and new file name)
#-------------------------------------
os.rename('test.txt' , 'newtest.txt')

#------------------------------
#to remove file use remove() method
#------------------------------
os.remove('newtest.txt')

#-----------------------------
# to create directory use mkdir() method
#--------------------------------
os.makedirs('firstdir')

#---------------------------------
# to move from dir to another dir
#---------------------------------
os.chdir('')
#-----------------------------------
# to print current dir working use getcwd() method
#-------------------------------------
os.getcwd()

#------------------------------------
#to remove dir use rmdir() method
#------------------------------------
os.rmdir('dirname')
```
